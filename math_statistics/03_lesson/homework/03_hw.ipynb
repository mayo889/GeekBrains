```python
import numpy as np
from math import sqrt, factorial

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

### Задание 1

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
arr = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])

mean = arr.sum() / arr.size

var_biased = sum((arr - mean) ** 2) / arr.size
var_unbiased = sum((arr - mean) ** 2) / (arr.size - 1)

std_biased = sqrt(var_biased)
std_unbiased = sqrt(var_unbiased)

print(f"Среднее: {mean}\n")

print(f"Смещенная оценка дисперсии: {round(var_biased, 3)}")
print(f"Несмещенная оценка дисперсии: {round(var_unbiased, 3)}\n")

print(f"Смещенное стандартное отклонение: {round(std_biased, 3)}")
print(f"Несмещенное стандартное отклонение: {round(std_unbiased, 3)}")
```

### Задание 2

В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
# возможны следующие 3 случая:
# из 1 ящика - 0 белых (2 других), из 2 ящика - 3 белых (1 другой)
# из 1 ящика - 1 белый (1 другой), из 2 ящика - 2 белых (2 других)
# из 1 ящика - 2 белых (0 других), из 2 ящика - 1 белый (3 других)
# Вероятности каждого случая складываем, так как они являются несовместимыми событиями

basket_1 = [0, 1, 2]
basket_2 = [3, 2, 1]
p_res = 0
for white_1, white_2 in zip(basket_1, basket_2):
    p_1 = combinations(5, white_1) * combinations(3, 2 - white_1) / combinations(8, 2)
    p_2 = combinations(5, white_2) * combinations(7, 4 - white_2) / combinations(12, 4)
    p_res += p_1 * p_2

print(f"Ответ: {round(p_res, 5)}")
```

### Задание 3

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
# Воспользуемся формулой Байеса

# Событие А - произошло попадание

# События B_i - выбрать одного из трех спортсменов (равновозможные события)
P_B_1 = 1 / 3
P_B_2 = 1 / 3
P_B_3 = 1 / 3

# Вероятности события А при наступлении события B_i (дано по условию)
P_A_B_1 = 0.9
P_A_B_2 = 0.8
P_A_B_3 = 0.6 

# Полная вероятность события А
P_A = P_B_1 * P_A_B_1 + P_B_2 * P_A_B_2 + P_B_3 * P_A_B_3

# Ответы по формуле Байеса
ans_1 = P_B_1 * P_A_B_1 / P_A
ans_2 = P_B_2 * P_A_B_2 / P_A
ans_3 = P_B_3 * P_A_B_3 / P_A

print("Вероятность, что выстрел в мишень был произведен:")
print(f"\t1 спортсменом: {round(ans_1, 3)} или {round(ans_1, 3) * 100}%")
print(f"\t2 спортсменом: {round(ans_2, 3)} или {round(ans_2, 3) * 100}%")
print(f"\t3 спорстменом: {round(ans_3, 3)} или {round(ans_3, 3) * 100}%")
```

### Задание 4

В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
# Воспользуемся формулой Байеса

# Событие А - студент сдал сессию

# События B_i - выбрать студента из факультета A, B, C
# Пусть на факультете А и В учится n студентов, тогда на факультете С 2n студентов
P_B_1 = 0.25 # n / 4n
P_B_2 = 0.25 # n / 4n
P_B_3 = 0.5 # 2n / 4n

# Вероятности события А при наступлении события B_i (дано по условию)
P_A_B_1 = 0.8
P_A_B_2 = 0.7
P_A_B_3 = 0.9 

# Полная вероятность события А
P_A = P_B_1 * P_A_B_1 + P_B_2 * P_A_B_2 + P_B_3 * P_A_B_3

# Ответы по формуле Байеса
ans_1 = P_B_1 * P_A_B_1 / P_A
ans_2 = P_B_2 * P_A_B_2 / P_A
ans_3 = P_B_3 * P_A_B_3 / P_A

print("Вероятность, что студент, сдавший сессию, учится:")
print(f"\tна факультете A: {round(ans_1, 3)} или {round(ans_1 * 100, 1)}%")
print(f"\tна факультете B: {round(ans_2, 3)} или {round(ans_2 * 100, 1)}%")
print(f"\tна факультете C: {round(ans_3, 3)} или {round(ans_3 * 100, 1)}%")
```

### Задание 5

Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
# Вероятность выйти из строя и не выйти для каждой детали
p1, q1 = 0.1, 0.9
p2, q2 = 0.2, 0.8
p3, q3 = 0.25, 0.75

print("Вероятность, что:")

res_1 = p1 * p2 * p3
print(f"\tвыйдут из строя все детали: {round(res_1, 3)} или {round(res_1 * 100, 1)}%")

res_2 = (q1 * p2 * p3) + (p1 * q2 * p3) + (p1 * p2 * q3)
print(f"\tвыйдут из строя только 2 детали: {round(res_2, 3)} или {round(res_2 * 100, 1)}%")

# Найдем через обратную вероятность, то есть из строя не выйдет ни одной детали
res_3 = 1 - q1 * q2 * q3
print(f"\tвыйдет из строя хотя бы 1 деталь: {round(res_3, 3)} или {round(res_3 * 100, 1)}%")

# Найдем через обратную вероятность, то есть из строя не выйдет ни одной детали
# или все 3 детали выйдут из строя
res_4 = 1 - q1 * q2 * q3 - p1 * p2 * p3
print(f"\tвыйдет из строя от одной до двух деталей: {round(res_4, 3)} или {round(res_4 * 100, 1)}%")
```
