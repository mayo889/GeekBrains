### Задание 2

О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

```python
D = 0.2
a = 0.5

# D = (b-a)^2 / 12  =>  b - a = +- sqrt(12*D)   => b = +- sqrt(12*D) + a
b_1 = sqrt(12 * D) + a
b_2 = -sqrt(12 * D) + a
b = max(a, b_1, b_2) # так как правая граница не может быть меньше левой

M = (a + b) / 2

print(f"Правая граница величины B: {round(b, 2)}")
print(f"Среднее значение величины B: {round(M, 2)}")
```

### Задание 4

Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

```python
def new_z(x, mu=174, std=8):
    return (x - mu) / std

# Так как распределение является нормальным, но не стандартным
# будем стандартизировать среднее и стандартное отклонение и переходить
# к соответствующим значениям стандартного распределения.
# По таблице накопленного нормального распределения будем находить долю общей площади
# и соответственно вероятность принадлежания роста какому-то отрезку значений

# Сразу заметим, что все величины роста отклоняются от среднего на величину кратную
# величине стандартного отклонения, поэтому заведем словарь табличных значений
# согласно таблице накопленного нормального распределения
z_dct = {-3.0: 0.0014, -2.0: 0.0228, -1.0: 0.1587,
         1.0: 0.8413, 2.0: 0.9772, 3.0: 0.9986}

print("Вероятность того,что случайным образом выбранный взрослый человек имеет рост")

# а). больше 182 см
z = new_z(182)
print(f"\tа) больше 182 см: {round(1 - z_dct[z], 3) * 100}%")

# б). больше 190 см
z = new_z(190)
print(f"\tа) больше 190 см: {round(1 - z_dct[z], 3) * 100}%")

# в). от 166 см до 190 см
z_1 = new_z(166)
z_2 = new_z(190)
print(f"\tв) от 166 см до 190 см: {round((z_dct[z_2] - z_dct[z_1]) * 100, 1)}%")

# г). от 166 см до 182 см
z_1 = new_z(166)
z_2 = new_z(182)
print(f"\tг) от 166 см до 182 см: {round((z_dct[z_2] - z_dct[z_1]) * 100, 1)}%")

# д). от 158 см до 190 см
z_1 = new_z(158)
z_2 = new_z(190)
print(f"\tд) от 158 см до 190 см: {round((z_dct[z_2] - z_dct[z_1]) * 100, 1)}%")

# е). не выше 150 см или не ниже 190 см
z_1 = new_z(150)
z_2 = new_z(190)
res = z_dct[z_1] + (1 - z_dct[z_2])
print(f"\tе) не выше 150 см или не ниже 190 см: {round(res * 100, 1)}%")

# ё) не выше 150 см или не ниже 198 см
z_1 = new_z(150)
z_2 = new_z(198)
res = z_dct[z_1] + (1 - z_dct[z_2])
print(f"\tё) не выше 150 см или не ниже 198 см: {round(res * 100, 1)}%")

# ж) ниже 166 см
z = new_z(166)
print(f"\tж) ниже 166 см: {round(z_dct[z], 3) * 100}%")
```

### Задание 5

На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
D = 25 # дисперсия
mu = 178 # мат ожидание
x = 190 # значение

std = sqrt(D) # стандартное отклонение

ans = (x - mu) / std

print(f"Ответ: отклоняется на {ans} сигмы")
```
