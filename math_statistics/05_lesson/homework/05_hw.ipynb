### Задание 1

Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

```python
mu = 80 # выборочная средняя
std = 16 # среднее квадратичное отклонение
n = 256 # объем выборки

# Так как доверительный интервал 0.95, то статистический уровень значимости равен 0.05
# Найдем по таблице накопленного нормального распределения значение критерия z при 0.95 + 0.05 / 2 = 0.975
z = 1.96

# воспользуемся следующей формулой, чтобы найти границы доверительного интервала
#  mu +- z(a/2) * sigma / sqrt(n)

left = mu - z * std / sqrt(n)
right = mu + z * std / sqrt(n)

print(f"Границы доверительного интервала с надежностью 95%: [{left};{right}]")
```

### Задание 2

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
x = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])
n = len(x) # число измерений - 10

# Посчитаем несмещенное среднеквадратичное отклонение, так как выборка мала
std = np.std(x, ddof=1)

# Воспользуемся критерием Стьюдента, так как нам не известна дисперсия величины X
v = n - 1 # число степений свободы - 9

# Так как доверительный интервал 0.95, то статистический уровень значимости равен 0.05
# Найдем по таблице значение критерия t при a / 2 = 0.05 / 2 = 0.025 и v = 9
t = 2.262

mu = x.mean() # среднее выборочное

# воспользуемся следующей формулой, чтобы найти границы доверительного интервала
#  mu +- t(a/2) * sigma / sqrt(n)

left = mu - t * std / sqrt(n)
right = mu + t * std / sqrt(n)

print(f"Границы доверительного интервала с надежностью 95%: [{round(left, 2)};{round(right, 2)}]")
```

### Задание 3

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв.мм.

```python
mu_0 = 17
alpha = 0.05
n = 100
mu = 17.5
var = 4

"""
Нулевая гипотеза Н0: mu = mu_0
Альтернативная гипотеза H1: mu > mu_0
"""
std = sqrt(var) # среднеквадратичное отклонение генеральной совокупности

# по таблице накопленного нормального распределения найдем z при (1 - 0.05 = 0.95)
z = 1.645

z_n = (mu - mu_0) / (std / sqrt(n))

if z_n > z:
    print(f"Принимаем альтернативную гипотезу H1, так как {z_n} > {z}")
else:
    print(f"Принимаем нулевую гипотезу H0, так как {z_n} < {z}")
```

### Задание 4

Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?

```python
x = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
mu_0 = 200
n = 10
interval = 0.99

"""
Нулевая гипотеза Н0: mu = mu_0
Альтернативная гипотеза H1: mu != mu_0
"""

alpha = 1 - interval # статистический уровень значимости = 0.01

# Воспользуемся критерием Стьюдента, так как нам не известна дисперсия величины X
v = n - 1 # число степений свободы - 9

# Найдем по таблице значение критерия t
# так как тест двусторонний
# при alpha / 2 = 0.01 / 2 = 0.005 
# и при v = 9 (число степений свободы)
t_left = -3.25 # левая граница критерия t нулевой гипотезы
t_right = 3.25 # правая граница критерия t нулевой гипотезы

mu = x.mean() # среднее выборочное
std = x.std(ddof=1) # среднеквадратичное отклонение выборки

# Найдем расчетное значение критерия Стьюдента
z_n = (mu - mu_0) / (std / sqrt(n))

if t_left < z_n < t_right:
    print(f"Принимаем нулевую гипотезу, так как расчетное значение t-критерия\
    {round(z_n, 2)} лежит в области принятия гипотезы: [{t_left};{t_right}]")
else:
    print(f"Отвергаем нулевую гипотезу, так как расчетное значение t-критерия\
    {round(z_n, 2)} НЕ лежит в области принятия гипотезы: [{t_left};{t_right}]")
```
